2. Quel est le nombre moyen de votes par pays ? Trier les pays par nombre moyen de votes décroissant.

```python
reponse2=db.movies.aggregate([
    { "$unwind": "$countries" },
    {"$group": {
            "_id": "$countries",
            "vote_moyen": {"$avg": "$imdb.votes"}
        }
    },
    {
        "$sort": {"vote_moyen" : -1}
    }
])

pd.DataFrame(list(reponse2))
#En moyenne au Bahamas les votes sont de 206061.
```

7. Quels sont les 5 acteurs les plus populaires (ceux qui ont joué dans le plus de films) ?

```python
reponse7=db.movies.aggregate([
        {
            "$unwind": "$cast"
        },
        {
            "$group": {
                "_id": "$cast",
                "nombre_films": {"$sum" :1}
            }
        },
        {
            "$sort": {"nombre_films": -1}
        },
        {
            "$limit": 5
        }
    ])

pd.DataFrame(list(reponse7))
#Un des 5 acteurs les plus populaires est Robert De Niro avec 60 films
```

8. Quelles sont les 6 années avec le plus de films produits ?

```python
reponse8=db.movies.aggregate(
    [
        {
            "$group": {
                "_id": "$year",
                "nombre_films": {"$sum": 1}
            }
        },
        {
            "$sort": {"nombre_films": -1}
        },
        {
            "$limit": 6
        }
    ]
)


pd.DataFrame(list(reponse8))
#En 2013, le nombre de films produit était de 1220
```

9. Combien de films ont été réalisés avant l'année 2000 et ont une note IMDb supérieure à 7 ?

```python
reponse9 = db.movies.aggregate(
    [
        {
            "$match":{
                "imdb.rating": {"$gt": 7},
                "year": {"$lt": 2000}
            }
        },
        {
            "$count": "nb_films"
        }
    ]
)

pd.DataFrame(list(reponse9))

#il y a 4747 films qui ont été réalisés avant l'année 2000 et qui ont une note IMDb supérieure à 7
```

10. Donner les 20 réalisateurs qui ont réalisé le plus de films (afficher seulement leur nom et le nombre de films).

```python
reponse10=db.movies.aggregate([
        {
            "$unwind": "$directors"
        },
        {
            "$group": {
                "_id": "$directors",
                "nombre_films": {"$sum" :1}
            }
        },
        {
            "$sort": {"nombre_films": -1}
        },
        {
            "$limit": 20
        }
    ])

pd.DataFrame(list(reponse10))
#Woody Allen a réalisé 40 films
```

11. Quelle est la note IMDb moyenne des films sortis après 2010 ?

```python
reponse11 = db.movies.aggregate(
    [
        {
            "$match": {
                 "year": {"$gt": 2010}
            }
        },
        {
            "$group": {
                "_id": None,
                "note_moyenne_apres_2010": {"$avg": "$imdb.rating"}
            }
        },
        {
            "$project": {  
                "_id": 0,
                "note_moyenne_apres_2010" : 1
            }
        }
    ]
)

pd.DataFrame(list(reponse11))

#la note imdb moyenne des films sortis après 2010 est de 6,54.
```

12. (bonus) Afficher les films dont le titre contient le mot "love".

```python
reponse12 = db.movies.aggregate([
    {
        "$match": {"title": {"$regex": "^love", "$options": "i"}}  
    },
    {
        "$project": {
            "_id": 0, 
            "title": 1 
        }
    }
])

pd.DataFrame(list(reponse12))
#Dans la base des films on peut en trouver plusieurs dont 1 intitulé "Love Finds Andy Hardy".
```
